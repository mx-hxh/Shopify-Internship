```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from functools import reduce
```

```python
df_pro = pd.read_csv('Property_Information_Data__Current_Calendar_Year_.csv', header = 0, engine = 'c')
df_pro = df_pro[df_pro['House Number'].notna() & df_pro['Street Name'].notna()]
df_pro.head()
```

```python
df_pro['Address'] = df_pro['House Number'].apply(str) + df_pro['Street Name']
df_pro = df_pro[df_pro['Actual Year Built']>=2015]
```

```python
df_group_ap = df_pro[['Address','Suite']].groupby('Address').agg('count').reset_index()
df_group_lat = df_pro[['Address','Latitude']].groupby('Address').agg('max').reset_index()
df_group_lon = df_pro[['Address','Longitude']].groupby('Address').agg('max').reset_index()
df_group_lot = df_pro[['Address','Lot Size']].groupby('Address').agg('max').reset_index()
df_group_year = df_pro[['Address','Actual Year Built']].groupby('Address').agg('max').reset_index()
df_group_nei = df_pro[['Address','Neighbourhood ID']].groupby('Address').agg('max').reset_index()
```

```python
data_frames = [df_group_ap, df_group_lat, df_group_lon, df_group_lot, df_group_year, df_group_nei]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Address'],
                                            how='outer'), data_frames)
df_merged['Apartment'] = df_merged['Suite'] > 0
df_year_high = df_merged[df_merged['Actual Year Built'].notna() & df_merged['Apartment']]
df_year_low = df_merged[df_merged['Actual Year Built'].notna() & df_merged['Apartment'] == False]
```

```python
year_diff = max(df_year_high['Actual Year Built']) - min(df_year_high['Actual Year Built'])
plt.hist(df_year_high['Actual Year Built'], bins = int(year_diff), rwidth = 0.8, align = 'mid')
plt.show()
```

```python
df_in = pd.read_csv('2016_Census_-_Population_by_Household_Income__Neighbourhood_Ward_.csv', header = 0, engine = 'c')
df_per = df_in[['Neighbourhood Number','Neighbourhood Name' ]]
names = df_in.columns
res = df_in[names[3:11]].sum(axis = 1)
low = df_in[names[3:4]].sum(axis = 1)
per = low / res
df_per['Low Income Rate'] = per
df_per['Low Income Rate'] = df_per['Low Income Rate'].fillna(0)
df_per['Total Household'] = df_in[names[3:12]].sum(axis = 1)
df_per
```

```python
df_year_high['Neighbourhood ID'].apply(int)
df_year_high
```

```python
df_high = pd.merge(df_year_high, df_per, left_on=  ['Neighbourhood ID'],
                   right_on= ['Neighbourhood Number'], 
                   how = 'left')
df_high.head(10)
```

```python
df_high_ad = df_high[['Neighbourhood ID','Apartment']].groupby('Neighbourhood ID').agg('count').reset_index()
df_high_rate = df_high[['Neighbourhood ID','Low Income Rate']].groupby('Neighbourhood ID').agg('max').reset_index()
df_high_su = df_high[['Neighbourhood ID','Suite']].groupby('Neighbourhood ID').agg('sum').reset_index()
df_high_total = df_high[['Neighbourhood ID','Total Household']].groupby('Neighbourhood ID').agg('max').reset_index()
dfs = [df_high_ad , df_high_rate, df_high_su, df_high_total]
df_merged_high = reduce(lambda  left,right: pd.merge(left,right,on=['Neighbourhood ID'],
                                            how='outer'), dfs)
```

```python
plt.scatter(x = df_merged_high['Low Income Rate'], y = df_merged_high['Apartment'])
plt.xlabel("Low Income Rate")
plt.ylabel("Number of New High Rise Building")
plt.title("Number of New High Rise Building Built by Neighbourhood 2015 - 2020")
plt.show()
```

```python
plt.scatter(x = df_merged_high['Low Income Rate'], y = df_merged_high['Total Household'], c = 'red')
plt.xlabel("Low Income Rate")
plt.ylabel("Number of Total Household")
plt.title("Number of Household by Neighbourhood 2016 Census")
plt.show()
```

```python
plt.scatter(x = df_merged_high['Low Income Rate'], y = df_merged_high['Suite'], c = 'g')
plt.xlabel("Low Income Rate")
plt.ylabel("Number of New Suites Avaliable")
plt.title("Number of New Suites Avaliable by Neighbourhood 2015 - 2020")
plt.show()
```

```python
df_low = pd.merge(df_year_low, df_per, left_on=  ['Neighbourhood ID'],
                   right_on= ['Neighbourhood Number'], 
                   how = 'left')
df_low_ad = df_low[['Neighbourhood ID','Apartment']].groupby('Neighbourhood ID').agg('count').reset_index()
df_low_rate = df_low[['Neighbourhood ID','Low Income Rate']].groupby('Neighbourhood ID').agg('max').reset_index()
df_low_total = df_low[['Neighbourhood ID','Total Household']].groupby('Neighbourhood ID').agg('max').reset_index()
dfs_low = [df_low_ad , df_low_rate, df_low_total]
df_merged_low = reduce(lambda  left,right: pd.merge(left,right,on=['Neighbourhood ID'],
                                            how='outer'), dfs_low)
df_merged_low 
```

```python
plt.scatter(x = df_merged_low['Low Income Rate'], y = df_merged_low['Apartment'])
plt.xlabel("Low Income Rate")
plt.ylabel("Number of New Houses")
plt.title("Number of New Houses Built by Neighbourhood 2015 - 2020")
plt.show()
```

```python
plt.scatter(x = df_merged_low['Low Income Rate'], y = df_merged_low['Total Household'], c = 'red')
plt.xlabel("Low Income Rate")
plt.ylabel("Number of Total Household")
plt.title("Number of Household by Neighbourhood 2016 Census")
plt.show()
```
